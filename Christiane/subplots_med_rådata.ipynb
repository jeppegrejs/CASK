```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## _Data for 20-års gennemsnit af temperatur gennem Holocæn_

The temperature reconstruction is based on an average of uplift corrected d18O data from Agassiz and Renland. This average has been corrected for changes in the d18O of seawater and calibrated to borehole temperature records from Camp Century, NGRIP, GRIP and DYE-3 (see Vinther et al., 2009 suppl. inf. for details).

```python
data_raw_holo = pd.read_excel("Vinther_etal_temp_20-kopi.xlsx", skiprows = 29).to_numpy() # importerer data, konverterer til et numpy-array
year_holo = data_raw_holo[:, 0] # b2k [yrs], starter i år 2000 og tæller baglæns med 20 års intervaller
temp_holo = data_raw_holo[:, 1] # deg ˚C
```

```python
# flipper y-akse data
year_array = np.array([])

for i in np.nditer([year_holo]):
    year_c = 2000 - i
    year_array = np.append(year_array, year_c)
```

```python
plt.plot(year_array, temp_holo, c = "red", label = "label")
plt.title("title")
plt.xlabel("age b2k [yrs]")
plt.ylabel("temp [degrees celsius]")
plt.xticks(rotation = 45)
plt.legend()
```

## _DMI temp data udsnit_

Det udsnit af DMI-temperaturmålinger emd daglige min/max temperaturere fra Illulisat, der anvendes til at udpakke temperaturdata fra Holocæn

```python
data_raw_DMI = pd.read_excel("DailyMinMax_Ilullisat_Cor-kopi.xlsx", skiprows = 0).to_numpy() # importerer data, konverterer til et numpy-array
year_DMI = data_raw_DMI[:, 0] # [yrs]
temp_DMI = data_raw_DMI[:, 1] # deg ˚C
```

```python
# Udvælger 20 år (1955 - 1975) og bruger deres årstidscyklus til udpakning af iskernedata

bottom = 1955
top = 1975
mask = (year_DMI >= bottom) & (year_DMI < top)

year_X = year_DMI[mask]
temp_X = temp_DMI[mask]
```

```python
plt.plot(year_X, temp_X, c = 'darkorange', label = "label")
plt.xlabel('Årstal')
plt.ylabel(r'Temperatur [$^\circ$C]')
plt.title("title")
plt.xticks(rotation = 45)
plt.legend()
```
